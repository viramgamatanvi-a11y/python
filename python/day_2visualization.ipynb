Stack Chart...

A stacked chart (bar or area) shows multiple data series stacked on top of one another, helping visualize the cumulative total and contribution of each series over time or categories..   

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
product_A = [5, 7, 6, 4, 5]
product_B = [3, 5, 2, 6, 4]
product_C = [2, 3, 4, 3, 2]
```

```python
plt.bar(days, product_A, label='Product A')
plt.bar(days, product_B, label='Product B')
plt.bar(days, product_C, label='Product C')

plt.legend()
plt.show()
```

```python
plt.bar(days,product_A,label='product_A',width=0.6)
plt.bar(days,product_B,label='product_B',width=0.4)
plt.bar(days,product_C,label='product_C',width=0.2)
plt.legend()
plt.show()
```

Horizontal chart....

```python
plt.barh(days, product_A, label='Product A')
plt.barh(days, product_B, label='Product B')
plt.barh(days, product_C, label='Product C')

plt.legend()
plt.show()
```

```python
# Stacked Area chart (Multiple line plots with fill -> Stacked Area chart)
```

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
product_A = [10, 20, 15, 25, 30]
product_B = [5, 15, 10, 20, 18]
product_C = [8, 12, 17, 10, 22]

plt.stackplot(
    days,
    product_A,
    product_B,
    product_C,
    labels=["product_A",'product_B','product_C']  
)

plt.legend()
plt.grid(axis='y',alpha=0.4)
plt.show()
```

Subplots....

subplots() allows you to display multiple plots (charts) in the same figure. Its especially useful for comparison....

```python
# Syntax
# ---------
# fig, axs = plt.subplots(nrows, ncols)
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y1, color='blue')
axs[0].set_title("Line Plot")

axs[1].bar(x, y2, color='green')
axs[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, y)
axs[0, 0].set_title("Plot 1")

axs[0, 1].bar(x, y)
axs[0, 1].set_title("Bar 1")

axs[1, 0].scatter(x, y)
axs[1, 0].set_title("Scatter 1")

axs[1, 1].barh(x, y)
axs[1, 1].set_title("Horizontal Bar")

plt.tight_layout()
plt.show()
```

Legends....

In Matplotlib, legends are used to label elements in your chart (like lines, bars, or points) so that the viewer understands what each represents...

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 1]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Legend Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()  # Show legend
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 1]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Legend Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# plt.legend()     # Show legend
plt.legend(loc='upper right')       # Position
plt.legend(fontsize=12)             # Font size
plt.legend(title='Lines')           # Title for the legend box
plt.legend(frameon=True)           # Remove box/frame
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 1]

plt.plot(x, y1, label='Line A')
plt.plot(x, y2, label='Line B')

plt.title('Legend Outside the Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Place legend outside
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()  # Adjust plot to not clip the legend

plt.show()
```

```python
#  Add Styling: Shadow, Title, Frame, Font Size
```

```python
plt.plot(x, y1, label='Line A')
plt.plot(x, y2, label='Line B')

plt.title('Legend Outside the Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend(
    loc='upper left',
    title='Legend Title',
    fontsize='medium',
    title_fontsize='large',
    shadow=True,
    frameon=True,
    facecolor='lightyellow',
    edgecolor='black'
)

plt.tight_layout()

plt.show()
```
