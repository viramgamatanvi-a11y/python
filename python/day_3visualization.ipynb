```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.Bar plot - Titanic Dataset

```python
# Load dataset

df=sns.load_dataset("titanic")

# Bar plot

plt.Figure(figsize=(8,5))
sns.barplot(x="class",y="fare",data=df,estimator=np.sum,errorbar=None)

plt.xlabel("Passenger class")
plt.ylabel("Total fare")
plt.title("Total fare by class")
plt.show()
```

```python
# Real-world styled dataset

df=pd.DataFrame({
    'Category' : ['Laptop','Laptop','Laptop','Phone','Phone','Tablet','Tablet','Tablet'],
    'Rating' : [4.5,4.0,4.2,4.8,4.6,4.1,4.0,3.9]
})

# Bar plot

sns.barplot(x='Category',y='Rating',data=df,errorbar=None,palette='deep',hue='Category')
plt.title('Average Customer Rating by Product Category')
plt.ylabel('Avg Rating')
plt.xlabel('Product Category')
plt.show()
```

```python
df=pd.DataFrame({
    'Category' : ['Laptop','Laptop','Phone','Phone','Tablet','Tablet'],
    'Gender' : ['Male','Female','Male','Female','Male','Female'],
    'Rating' : [4.5,4.1,4.7,4.6,4.0,4.1]
})

sns.barplot(x='Category',y='Rating',data=df,palette='Set2',errorbar=None,hue='Gender')
plt.title('Avg RAting by Product & Gender')
plt.show()
```

2.Line plot - Flights Dataset

```python
flights=sns.load_dataset("flights")

line_data=flights.groupby("year") ['passengers'].sum().reset_index()

# Line _data

plt.Figure(figsize=(10,5))
sns.lineplot(data=line_data,x='year',y='passengers',marker='o')
plt.title("Total passengers per year")
plt.show()
```

3.Scater plot with Regression line-tips dataset

```python
tips=sns.load_dataset("tips")

plt.Figure(figsize=(8,5))

# sns.scatterplot(data=tips,x='total_bill',y='tip',hue="time")

sns.regplot(data=tips,x="total_bill",y='tip',scatter=True,color='Black')
plt.title('Total Bill us Tip with Regression line')
plt.show()
```

4.Pie chart

```python
pie_data = tips.groupby("day", observed=False)["total_bill"].sum()

plt.figure(figsize=(7,7))
plt.pie(
    pie_data,
    labels=pie_data.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Total Bill Distribution by Day (Pie Chart)")
plt.show()
```

5.Histogram - Titanic Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset
df = pd.read_csv("titanic.csv")

# Step 2: Plot age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution on Titanic")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

6.Countm plot - Titanic Dataset

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='class',hue='sex')

plt.title("Passenger count by class and Gender")
plt.show()
```

7.Heatmap - correlation in Titanic Dataset 

```python
df=sns.load_dataset('tips')
carr=df.corr(numeric_only=True)

# corr

plt.Figure(figsize=(10,6))

sns.heatmap(carr,annot=True,linewidths=0.5,cmap='YlOrRd')

plt.title("Correlation Heatmap-Titanic")
plt.show()
```
