->3.Scatter plot

```python
df = sns.load_dataset("titanic")

line_data = df.groupby("pclass")['fare'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.scatter(line_data['pclass'], line_data['fare'])

plt.title("Total Fare per Class")
plt.xlabel("Passenger Class")
plt.ylabel("Total Fare")

plt.show()
```

->4.Pie plot

```python
df = sns.load_dataset("titanic")

pie_data = df.groupby("pclass")['fare'].sum().reset_index()

plt.figure(figsize=(7,7))
plt.pie(pie_data['fare'], labels=pie_data['pclass'], autopct="%0.1f%%")

plt.title("Total Fare Contribution per Class")
plt.show()
```

->5.Histogram plot

```python
df = sns.load_dataset("titanic")

plt.figure(figsize=(10,5))
plt.hist(df['fare'], bins=20, color='skyblue', edgecolor='black')

plt.title("Fare Distribution of Passengers")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```

->6.Heatmap plot

```python
df = sns.load_dataset("titanic")
heat_data = df.pivot_table(values='fare', index='pclass', columns='sex', aggfunc='sum')
plt.figure(figsize=(8,5))
sns.heatmap(heat_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Total Fare by Class and Gender")
plt.xlabel("Gender")
plt.ylabel("Passenger Class")

plt.show()
```
