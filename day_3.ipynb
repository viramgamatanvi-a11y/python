Boolean Mask (Masking Array)

```python
arr
```

```python
l
```

```python
print(np.ma.array(arr,mask=l))
```

```python
odd_number=np.ma.array(arr,mask=arr%2==0)
print(odd_number)
```

2.Vector Broadcasting

1.Row-vactor
2.Column_vactor

1.Row_Vactor

```python
a1=np.array([[1,2,3],[4,5,6]])
a2=np.array([10,20,30])
```

```python
print(a1+a2)
```

2.Column-Vactor

```python
a1=np.array([[1,2,3],[4,5,6]])
a2=np.array([10,20]).reshape(-1,1)
```

```python
print(a1+a2)
```

3.Matrix Broadcasting

```python
a1=np.array([[1],[2]])
a2=np.array([[10,20,30]])

print(a1)
print()
print(a2)
print()

a3=a1+a2
print(a3)
```
