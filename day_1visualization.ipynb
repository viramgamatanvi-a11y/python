Bar chart....

Bar chart is used for comparision......

Bar chart = plt.bar

```python
x=['Math','Science','English']
y=[85,90,70]
```

```python
plt.bar(x,y)
```

```python
plt.bar(x,y)

plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,width=0.5)     # width must be in range of 0.0 to 1.0
plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,color="green",width=0.3)
plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,color=['red','green','blue'])
plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,alpha=0.5)        # alphan is opacity, ranges from 0.0 to 1.0
plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.Figure(figsize=(4,2))     # Figsize is in inches

plt.bar(x,y,align="center")       # align can be 'center' or 'edge'ArithmeticError

plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,color="tomato", edgecolor="blue", linewidth=2)

plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(x,y,color="tomato",edgecolor="black", linewidth=2)
```

```python
# Optional : add value label on top of bars
```

```python
plt.bar(x,y,color="tomato", edgecolor="blue", linewidth=2)

plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")

plt.text(0,85+1,str(85),ha="center",fontweight="bold")
plt.text(1,90+1,str(90),ha="center",fontweight="bold")
plt.text(2,70+1,str(70),ha="center",fontweight="bold")

plt.show()
```

```python
plt.bar(x,y,color="tomato", edgecolor="blue", linewidth=2)

plt.title("Subject-wise Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")

for i, v in enumerate(y):
    plt.text(i,v+1,str(v),ha="center",fontweight="bold")
    
plt.tight_layout()       # adjust layout to prevent clipping of tick-labels

plt.show()
```

Horizontal Bar chart....

left->right...

Syntax : plt.barh()

```python
plt.barh(x, y, color="tomato", edgecolor="blue", linewidth=2)

plt.title("Subject-wise Marks")
plt.xlabel("Marks")
plt.ylabel("Subject")

for i, v in enumerate(y):
    plt.text(v + 1, i, str(v), va="center", fontweight="bold")   #  Correct

plt.tight_layout()
plt.show()
```

Line chart....

Line chart used for : A line chart shows data trends over time , visualize changes, growth, or decline...

Syntax : plt.plot()

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','oct','Nov','dec']
revenue=[5000, 5200, 5800, 6200, 6000, 6700, 7100, 6900, 7200, 7500, 7700, 8000]
```

```python
plt.plot(months,revenue)

plt.title("Monthly Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.plot(months,revenue,marker="*")
plt.show()
```

```python
plt.plot(months,revenue,marker="o")
plt.show()
```

```python
plt.plot(months,revenue,marker="+")
plt.show()
```

```python
plt.plot(months,revenue,marker="o",markersize=8,color="black",markerfacecolor='red')
plt.show()
```

```python
plt.Figure(figsize=(12,8))

plt.plot(months,revenue,marker="o",markersize=8,color="black",markerfacecolor='red', linewidth=2)

for i,val in enumerate(revenue):
    plt.text(i, val+120 , str(val), ha="center", fontweight="bold")
    
plt.grid(True,linestyle='--',alpha=0.7,axis="both")

plt.savefig('monthly_revenue.png',dpi=300)

plt.show()
```

Scatter chart....

A scatter plot is used to display the relationship between two numeric variables...

Syntax : plt.scatter()

```python
ad_spend = [5, 6, 7, 6, 8, 7.5, 9, 10, 9.5, 11, 10.5, 12]
sales = [40, 45, 50, 46, 54, 53, 58, 62, 59, 65, 64, 70]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']


plt.scatter(ad_spend,sales)
plt.show()
```

```python
plt.Figure(figsize=(10,8))
plt.scatter(ad_spend,sales,color='orange',edgecolors='black',s=100)   # s is size of the marker
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(ad_spend,sales,color='orange',edgecolors='black',s=100)

plt.xlabel('Ad Spend')
plt.ylabel("Sales")

for i, m in enumerate(months):
    plt.text(ad_spend[i], sales[i]+1.5, m, ha="center",fontweight='bold')
    
plt.show()
```

Pie Chart....


A Pie Chart shows percentage or proportion of different categories as slices of a circle...

Syntax : plt.pie()

```python
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels)
plt.show()
```

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]

plt.pie(sizes, labels=labels)
plt.title('Fruit Sales Distribution')
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 40, 10, 30]
colors = ['red', 'yellow', 'pink', 'brown']

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%.1f%%',
    startangle=90
)
plt.title('Fruit Market Share')
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 50, 10, 25]
explode = (0, 0.1, 0, 0)  # Only Banana pops out!

plt.pie(
    sizes,
    labels=labels,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)
plt.title('Fruit Popularity')
plt.show()
```

```python
products = ['Laptop', 'Tablet', 'Smartphone', 'Accessories']
sales = [400, 150, 300, 150]

# Create Pie
plt.pie(
    sales,
    labels=products,
    autopct='%1.0f%%',
    startangle=140,
    colors=['#4CAF50', '#FFC107', '#2196F3', '#9C27B0']
)

plt.title('Product Sales Distribution')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Fruit Distribution')
plt.axis('equal')  # Equal aspect ratio ensures circular pie
plt.show()
```

```python
sizes = [25, 35, 20, 20]
labels = ['Dogs', 'Cats', 'Birds', 'Fish']
explode = (0.1, 0, 0, 0)  # Only explode the 1st slice
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Pet Ownership')
plt.show()
```

```python
data = [12, 22, 18, 32, 16]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

plt.pie(data, labels=labels, colors=colors,
        autopct=lambda p: '{:.0f}%'.format(p) if p > 5 else '',
        pctdistance=0.85, startangle=140,
        wedgeprops={'linewidth': 2, 'edgecolor': 'white'},
        textprops={'fontsize': 12, 'color': 'black'})

# Draw circle to make it a donut
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Customized Pie Chart')
plt.axis('equal')
plt.show()
```

```python
# Data
size_outer = [20, 30, 10, 40]
size_inner = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
colors_outer = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
colors_inner = ['lightgray'] * 10

# Outer pie
plt.pie(size_outer, radius=1.3, colors=colors_outer,
        wedgeprops=dict(width=0.3, edgecolor='w'))

# Inner pie
plt.pie(size_inner, radius=1.3-0.3, colors=colors_inner,
        wedgeprops=dict(width=0.3, edgecolor='w'))

plt.title('Nested Pie Chart')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 30, 25, 25]

# Create Pie Chart
plt.pie(sizes, labels=labels, autopct='%.1f%%', startangle=90)

# Draw a white circle in center to make it a Donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Title
plt.title('Fruit Sales - Donut Chart')

# Show
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Laptop', 'Tablet', 'Phone', 'Accessories']
sizes = [50, 20, 20, 10]
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']
explode = (0.05, 0.05, 0.05, 0.05)  # All slices slightly out

# Create Pie
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    shadow=True
)

# Center Circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Product Sales Distribution - Donut Chart')
plt.show()
```

Histogram Chart...

A Histogram is a type of bar chart that represents the distribution of numerical data...

```python
import matplotlib.pyplot as plt

# Sample Data: Math exam scores
scores = [45, 78, 56, 89, 90, 67, 80, 56, 92, 45,
          60, 75, 85, 55, 70, 90, 88, 72, 60, 65,
          68, 77, 85, 95, 48, 52, 60, 75, 80, 82]

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(scores, bins=7, color='skyblue', edgecolor='black', rwidth=0.9)

# Adding labels and title
plt.title('Distribution of Math Exam Scores', fontsize=16)
plt.xlabel('Score Range', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# Display grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample Data
math_scores =    [45, 78, 56, 89, 90, 67, 80, 56, 92, 45,
                  60, 75, 85, 55, 70, 90, 88, 72, 60, 65,
                  68, 77, 85, 95, 48, 52, 60, 75, 80, 82]

science_scores = [40, 70, 60, 85, 88, 66, 75, 50, 90, 42,
                  58, 72, 80, 50, 65, 88, 86, 70, 55, 60,
                  63, 74, 83, 91, 45, 51, 58, 73, 78, 79]
```

```python
plt.figure(figsize=(10, 6))

# Plot Math Scores
plt.hist(math_scores, bins=7, color='blue', alpha=0.5, label='Math', edgecolor='black')

# Plot Science Scores
plt.hist(science_scores, bins=7, color='green', alpha=0.5, label='Science', edgecolor='black')

# Title and Labels
plt.title('Score Distribution: Math vs Science', fontsize=16)
plt.xlabel('Score Range', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# Legend and Grid
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
